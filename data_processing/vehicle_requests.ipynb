```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read the CSV file
df = pd.read_csv('../pcd_divider/modified_file.csv')

# Step 2: Count distinct file names by Vehicle ID
distinct_file_counts = df.groupby('Vehicle ID')['Tile Name'].nunique()

# Step 3: Plot the graph as a line graph
vehicle_ids = list(distinct_file_counts.index)
file_counts = list(distinct_file_counts.values)

plt.plot(vehicle_ids, file_counts, marker='o', linestyle='-', color='b')
plt.xlabel('Vehicle ID')
plt.ylabel('Number of Distinct File Names')
plt.title('Vehicle ID vs Number of Distinct File Names')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# List of CSV files with different grid sizes
csv_files = [
    '/home/sayon/Thesis/static_pcd_divider/processing/64_validity_with_size.csv',
    '/home/sayon/Thesis/static_pcd_divider/processing/32_validity_with_size.csv',
    '/home/sayon/Thesis/static_pcd_divider/processing/16_validity_with_size.csv',
    '/home/sayon/Thesis/static_pcd_divider/processing/8_validity_with_size.csv',
    '/home/sayon/Thesis/static_pcd_divider/processing/4_validity_with_size.csv',
    '/home/sayon/Thesis/pcd_cache/validity_with_size.csv'
]

# Step 1: Prepare data for the bar graph
data = []

# Step 2: Read and process each CSV file separately
for csv_file in csv_files:
    # Read the CSV file
    df = pd.read_csv(csv_file)

    # Count distinct file names by Vehicle ID
    distinct_file_counts = df.groupby('Vehicle ID')['Tile Name'].nunique()

    # Append data for the current file
    file_data = {
        'File': csv_file,
        'Vehicle ID': list(distinct_file_counts.index),
        'Distinct File Count': list(distinct_file_counts.values)
    }
    data.append(file_data)

# Step 3: Plot the bar graph
plt.figure(figsize=(10, 6))  # Set the figure size

bar_width = 0.15  # Width of each bar
index = range(len(data[0]['Vehicle ID']))  # Index for the x-axis

for i, file_data in enumerate(data):
    # Extract data for the current file
    file_name = file_data['File']
    vehicle_ids = file_data['Vehicle ID']
    file_counts = file_data['Distinct File Count']

    # Calculate the position for each bar graph
    pos = [j + i * bar_width for j in index]

    # Plot the bar graph for the current file
    plt.bar(pos, file_counts, bar_width, label=f'File: {file_name}')

plt.xlabel('Vehicle ID')
plt.ylabel('Number of Distinct File Names')
plt.title('Vehicle ID vs Number of Distinct File Names')
plt.legend()
plt.xticks([r + (len(data) * bar_width) / 2 for r in index], data[0]['Vehicle ID'])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# List of CSV files with different grid sizes
csv_files = [
    '/home/sayon/Thesis/static_pcd_divider/processing/50_mbps/64_validity_with_size.csv',
    '/home/sayon/Thesis/static_pcd_divider/processing/50_mbps/32_validity_with_size.csv',
    '/home/sayon/Thesis/static_pcd_divider/processing/50_mbps/16_validity_with_size.csv',
    '/home/sayon/Thesis/static_pcd_divider/processing/50_mbps/8_validity_with_size.csv',
    '/home/sayon/Thesis/static_pcd_divider/processing/50_mbps/4_validity_with_size.csv',
    '/home/sayon/Thesis/pcd_cache/validity_with_size.csv'
]

# Specify the legend labels
legend_labels = [
    '64x64 grid',
    '32x32 grid',
    '16x16 grid',
    '8x8 grid',
    '4x4 grid',
    'non-uniform grid'
]

# Step 1: Prepare data for the bar graph
data = []

# Step 2: Read and process each CSV file separately
for i, csv_file in enumerate(csv_files):
    # Read the CSV file
    df = pd.read_csv(csv_file)

    # Filter data for Vehicle IDs 1 to 10
    filtered_df = df[df['Vehicle ID'].between(1, 10)]

    # Count distinct file names by Vehicle ID
    distinct_file_counts = filtered_df.groupby('Vehicle ID')['Tile Name'].nunique()

    # Append data for the current file
    file_data = {
        'File': csv_file,
        'Vehicle ID': list(distinct_file_counts.index),
        'Distinct File Count': list(distinct_file_counts.values),
        'Legend Label': legend_labels[i]
    }
    data.append(file_data)

# Step 3: Plot the bar graph
plt.figure(figsize=(10, 6))  # Set the figure size

bar_width = 0.15  # Width of each bar
index = range(1, 11)  # Index for the x-axis, limited to Vehicle IDs 1 to 10

for i, file_data in enumerate(data):
    # Extract data for the current file
    file_name = file_data['File']
    vehicle_ids = file_data['Vehicle ID']
    file_counts = file_data['Distinct File Count']
    legend_label = file_data['Legend Label']

    # Calculate the position for each group of bars
    pos = [j + i * bar_width for j in index]

    # Plot the bar graph for the current file
    plt.bar(pos, file_counts, bar_width, label=legend_label)

plt.xlabel('Vehicle ID')
plt.ylabel('Number of Requests')
plt.title('Vehicle ID vs Number of Requests')
plt.legend()

# Increase the legend font size
plt.legend(prop={'size': 12}) 

plt.xticks([r + (len(data) * bar_width) / 2 for r in index], index)
plt.savefig('requests.eps', format='eps')

plt.show()
```
