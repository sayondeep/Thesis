```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read the CSV file
df = pd.read_csv('../pcd_divider/modified_file.csv')

# Step 2: Calculate file sizes by Vehicle ID
file_sizes = {}
for index, row in df.iterrows():
    vehicle_id = row['Vehicle ID']
    tile_name = row['Tile Name']
    file_size = os.path.getsize(tile_name) / (1024 * 1024)  # Convert to MB
    if vehicle_id in file_sizes:
        file_sizes[vehicle_id] += file_size
    else:
        file_sizes[vehicle_id] = file_size

# Step 3: Plot the graph
vehicle_ids = list(file_sizes.keys())
file_sizes = list(file_sizes.values())

plt.bar(vehicle_ids, file_sizes)
plt.xlabel('Vehicle ID')
plt.ylabel('File Size (MB)')
plt.title('Vehicle ID vs File Size')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read the CSV file
df = pd.read_csv('../pcd_divider/modified_file.csv')

# Step 2: Count distinct file names by Vehicle ID
distinct_file_counts = df.groupby('Vehicle ID')['Tile Name'].nunique()

# Step 3: Plot the graph
vehicle_ids = list(distinct_file_counts.index)
file_counts = list(distinct_file_counts.values)

plt.bar(vehicle_ids, file_counts)
plt.xlabel('Vehicle ID')
plt.ylabel('Number of requests')
plt.title('Vehicle ID vs Number of Distinct File Names')
plt.show()
```
