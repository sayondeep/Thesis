```python
import matplotlib.pyplot as plt
hits =[8370, 8430, 8435, 8504, 8538, 8577, 8581, 8587, 8603, 8606, 8606, 8606, 8606, 8606]
miss =[1440, 1380, 1375, 1306, 1272, 1233, 1229, 1223, 1207, 1204, 1204, 1204, 1204, 1204]



cache_sizes = range(70, 201, 10)

# Plotting hits
plt.figure(1)
plt.plot(cache_sizes, hits)
plt.xlabel('Cache Size (MB)')
plt.ylabel('Hits')
plt.title('Cache Hits')

# Plotting misses
plt.figure(2)
plt.plot(cache_sizes, miss)
plt.xlabel('Cache Size (MB)')
plt.ylabel('Misses')
plt.title('Cache Misses')

plt.show()
```
