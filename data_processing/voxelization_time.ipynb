```python
import pandas as pd
import matplotlib.pyplot as plt

# List of CSV files
csv_files = ['/home/sayon/Thesis/pcd_cache/time_to_voxelize.csv',
             '/home/sayon/Thesis/voxelization/64_time_to_voxelize.csv',
             '/home/sayon/Thesis/voxelization/32_time_to_voxelize.csv',
             '/home/sayon/Thesis/voxelization/16_time_to_voxelize.csv',
             '/home/sayon/Thesis/voxelization/8_time_to_voxelize.csv',
             '/home/sayon/Thesis/voxelization/4_time_to_voxelize.csv']

# Specify x labels manually
x_labels = ['Non-uniform', '64x64', '32x32', '16x16', '8x8', '4x4']

# Initialize total times for each file
total_times = []

# Calculate total time for each file
for csv_file in csv_files:
    df = pd.read_csv(csv_file)
    total_time = 0
    for index, row in df.iterrows():
        if row['Time'] == row['Valid_From']:
            file_time = row['time_to_voxelize']
            total_time += file_time
    total_times.append(round(total_time, 2))  # Restrict to 2 decimal places

# Plot the graph
plt.bar(x_labels, total_times)
plt.xlabel('CSV File')
plt.ylabel('Total time (s)')
plt.title('Total time for voxelization')
plt.xticks(rotation=45)

# Add labels on top of the bars
for i, v in enumerate(total_times):
    plt.text(i, v, str(v), ha='center', va='bottom')

plt.show()
```
