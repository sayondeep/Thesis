```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read the CSV file
df = pd.read_csv('/home/sayon/Thesis/pcd_cache/validity_with_size.csv')

# Step 2: Calculate file sizes by Vehicle ID
file_sizes = {}
for index, row in df.iterrows():
    vehicle_id = row['Vehicle ID']
    tile_name = row['Tile Name']
    is_updated = row['Is_Update']
    valid_from = row['Valid_From']
    time = row['Time']

    # Check if is_updated is 1 and valid_from matches the time
    if is_updated == 1 and valid_from == time:
        file_size = os.path.getsize(tile_name) / (1024 * 1024)  # Convert to MB

        if vehicle_id in file_sizes:
            file_sizes[vehicle_id] += file_size
        else:
            file_sizes[vehicle_id] = file_size

# Step 3: Plot the graph
vehicle_ids = list(file_sizes.keys())
file_sizes = list(file_sizes.values())

plt.bar(vehicle_ids, file_sizes)
plt.xlabel('Vehicle ID')
plt.ylabel('File Size (MB)')
plt.title('Vehicle ID vs File Size')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# List of CSV files
csv_files = ['/home/sayon/Thesis/static_pcd_divider/processing/64_validity_with_size.csv', '/home/sayon/Thesis/static_pcd_divider/processing/32_validity_with_size.csv', '/home/sayon/Thesis/static_pcd_divider/processing/16_validity_with_size.csv', '/home/sayon/Thesis/static_pcd_divider/processing/8_validity_with_size.csv', '/home/sayon/Thesis/static_pcd_divider/processing/4_validity_with_size.csv', '/home/sayon/Thesis/pcd_cache/validity_with_size.csv']

# Initialize total sizes for each file
total_sizes = []

# Calculate total size for each file
for csv_file in csv_files:
    df = pd.read_csv(csv_file)
    total_size = 0
    for index, row in df.iterrows():
        if row['Time'] == row['Valid_From']:
            file_size = row['Size']
            total_size += file_size
    total_sizes.append(total_size)

# Plot the graph
plt.bar(csv_files, total_sizes)
plt.xlabel('CSV File')
plt.ylabel('Total Size')
plt.title('Total Size of Files')
plt.xticks(rotation=45)
plt.show()
```
